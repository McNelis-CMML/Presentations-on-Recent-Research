# Small Open Economy DSGE Model

## Overview

This is a Dynamic Stochastic General Equilibrium (DSGE) model of a small open economy featuring:

- **CES production function** with capital, labor, and imported oil inputs
- **Q-theory of investment** with capital adjustment costs
- **Current account dynamics** with endogenous net foreign assets
- **Sticky prices** via Calvo pricing mechanism
- **Monetary policy** via Taylor rule with interest rate smoothing
- **Exchange rate dynamics** via uncovered interest parity
- **Two exogenous shocks**: Total Factor Productivity (TFP) and oil price shocks

The model is designed to analyze the transmission of productivity and oil price shocks through the domestic economy, considering both real and nominal rigidities, as well as international financial linkages.

- **$\gamma$**: Labor supply scaling factor (calibrated from steady-state labor market equilibrium)

## Key Model Equations

### 1. Production Technology

**CES Production Function:**

$$\text{ymarg}_t = \alpha_K k_{t-1}^{\rho_c} + \alpha_L l_t^{\rho_c} + \alpha_O \text{oil}_t^{\rho_c}$$

$$\hat{y}_t = (\text{ymarg}_t)^{1/\rho_c}$$

$$y_t = \bar{A} \cdot e^{a_t} \cdot \hat{y}_t$$

The production function exhibits constant elasticity of substitution between capital, labor, and oil. The parameter $\rho_c$ determines the degree of substitutability:
- $\rho_c \to 0$: Cobb-Douglas
- $\rho_c = -0.1$: Inputs are **less substitutable** than Cobb-Douglas (complements)

The elasticity of substitution is: $\sigma_{CES} = \frac{1}{1-\rho_c} = \frac{1}{1.1} \approx 0.91$

**Marginal Products:**

$$\text{mpk}_t = \bar{A} \cdot e^{a_t} \cdot (\text{ymarg}_t)^{1/\rho_c - 1} \cdot \alpha_K \cdot k_{t-1}^{\rho_c - 1}$$

$$w_t = \bar{A} \cdot e^{a_t} \cdot (\text{ymarg}_t)^{1/\rho_c - 1} \cdot \alpha_L \cdot l_t^{\rho_c - 1}$$

$$\text{mpoil}_t = \bar{A} \cdot e^{a_t} \cdot (\text{ymarg}_t)^{1/\rho_c - 1} \cdot \alpha_O \cdot \text{oil}_t^{\rho_c - 1}$$

### 2. Factor Demand

**Oil demand condition:**

$$\frac{s_t}{p_t} \cdot \bar{P}^{oil} \cdot e^{a_t^{oil}} = \text{mpoil}_t$$

The real exchange rate-adjusted price of oil equals its marginal product.

**Labor supply:**

$$l_t = \left(\frac{w_{t-1}(1-\tau_w)\lambda_t}{\gamma}\right)^{\phi_L}$$

Labor supply depends on the after-tax real wage, weighted by marginal utility, with Frisch elasticity $\phi_L = 0.5$.

### 3. Capital Accumulation and Investment

**Capital law of motion:**

$$k_t = (1 - \delta)k_{t-1} + i_t$$

where $\delta = 0.04$ is the depreciation rate.

**Investment with adjustment costs (Q-theory):**

$$i_t = \delta k_{t-1} + \frac{(q_t - 1)}{\Phi} k_{t-1}$$

When Tobin's Q exceeds 1, firms invest more than depreciation. The parameter $\Phi = 10$ governs adjustment cost magnitude.

**Adjustment cost derivative:**

$$C_K(k_{t-1}, i_t) = \frac{\Phi}{2}\left(\frac{i_t - \delta k_{t-1}}{k_{t-1}}\right)^2 - \Phi\left(\frac{i_t - \delta k_{t-1}}{k_{t-1}}\right)\left(\frac{i_t}{k_{t-1}}\right)$$

### 4. Asset Pricing

**Tobin's Q equation:**

$$q_t = \beta \left[\text{mpk}_{t+1} + (1-\delta)q_{t+1} - C_K(k_t, i_{t+1})\right]$$

The value of installed capital equals the discounted sum of marginal product, continuation value, and adjustment costs.

**Rental rate of capital:**

$$r_t^k = mc_t \cdot \text{mpk}_t$$

### 5. Household Optimization

**Marginal utility:**

$$\lambda_t = \frac{c_t^{-\sigma_c}}{1 + \tau_c}$$

where $\sigma_c = 2$ is the coefficient of relative risk aversion.

**Euler equation:**

$$\lambda_t = \beta(1 + r_t^{dom})\lambda_{t+1}$$

Households smooth consumption over time based on the domestic nominal interest rate.

### 6. Current Account and Net Foreign Assets

**Current account identity:**

$$\text{nfa}_t = (1+r_t)\text{nfa}_{t-1} + \left[y_t - (1+\tau_c)c_t - i_t - \frac{s_t}{p_t}\bar{P}^{oil}e^{a_t^{oil}}\text{oil}_t - \tau_w w_t l_t - \tau_k r_t^k k_{t-1}\right]$$

Trade balance = Output - Consumption - Investment - Oil imports - Tax revenues

**Debt-elastic interest rate:**

$$r_t = \bar{r} - \Phi_B \cdot \text{nfa}_{t-1}$$

When the country is a net debtor ($\text{nfa} < 0$), it faces a higher interest rate premium. The parameter $\Phi_B = 0.001$ governs the elasticity.

### 7. Sticky Prices (Calvo Pricing)

**Marginal cost:**

$$mc_t = \frac{1}{\bar{A} e^{a_t}} \left[\alpha_L^{\sigma_{mc}} w_t^{1-\sigma_{mc}} + \alpha_K^{\sigma_{mc}} \text{mpk}_t^{1-\sigma_{mc}} + \alpha_O^{\sigma_{mc}} \left(\bar{P}^{oil}e^{a_t^{oil}}\right)^{1-\sigma_{mc}}\right]^{\frac{1}{1-\sigma_{mc}}}$$

where $\sigma_{mc} = \frac{1}{1-\rho_c}$ is the elasticity of substitution in the cost function.

**New Keynesian Phillips Curve:**

$$\pi_t = \beta \pi_{t+1} + \frac{(1-\theta)(1-\beta\theta)}{\theta} \log\left(\frac{mc_{t-1}}{\overline{mc}}\right)$$

Inflation depends on expected future inflation and the deviation of marginal cost from steady state. With $\theta = 0.75$, firms adjust prices on average every 4 periods.

**Price level evolution:**

$$p_t = p_{t-1} + \pi_t$$

### 8. Monetary Policy and Exchange Rate

**Taylor Rule with ZLB:**

$$r_t^{dom} = \max\left\{0, \, (1-\rho_r)\bar{r} + \rho_r r_{t-1}^{dom} + (1-\rho_r)\left[\phi_\pi \pi_{t+1} + \phi_y \log\left(\frac{y_{t+1}}{\bar{y}}\right)\right]\right\}$$

The central bank responds to expected inflation ($\phi_\pi = 1.5$) and the output gap ($\phi_y = 0.5$), with interest rate smoothing ($\rho_r = 0.9$). The max operator imposes the zero lower bound.

**Uncovered Interest Parity (UIP):**

$$r_t^{dom} - r_t = s_{t+1} - s_t$$

The interest rate differential equals the expected depreciation rate of the nominal exchange rate.

### 9. Exogenous Shocks

**TFP shock:**

$$a_t = \rho_a a_{t-1} + \sigma_a \varepsilon_t^a$$

**Oil price shock:**

$$a_t^{oil} = \rho_{oil} a_{t-1}^{oil} + \sigma_{oil} \varepsilon_t^{oil}$$

Both shocks follow AR(1) processes with high persistence ($\rho_a = \rho_{oil} = 0.9$) and standard deviations $\sigma_a = \sigma_{oil} = 0.005$.

### Key Steady-State Relationships

**Discount factor from Euler equation:**
$$\beta = \frac{1}{\overline{\text{mpk}} + (1-\delta)}$$

This ensures that the household's intertemporal consumption decision is consistent with the return to capital.

**Consumption from resource constraint:**
$$\bar{c} = \frac{1 - \bar{o} \cdot \overline{\text{mpoil}} - \delta\bar{k} - \tau_w \cdot \overline{w} \cdot \bar{l} - \tau_k \cdot \bar{k} \cdot \overline{\text{mpk}}}{1 + \tau_c}$$

Consumption is determined residually after accounting for:
- Oil imports: $\bar{o} \cdot \overline{\text{mpoil}}$
- Depreciation: $\delta\bar{k}$
- Tax revenues: $\tau_w \overline{w} \bar{l} + \tau_k \bar{k} \overline{\text{mpk}}$

**CES production calibration:**
$$\overline{\text{ymarg}} = \alpha_K \bar{k}^{\rho_c} + \alpha_L \bar{l}^{\rho_c} + \alpha_O \bar{o}^{\rho_c}$$

$$\bar{\hat{y}} = (\overline{\text{ymarg}})^{1/\rho_c}$$

$$\bar{A} = \frac{\bar{y}}{\bar{\hat{y}}}$$

The TFP scaling factor $\bar{A}$ is chosen to normalize output to 1.

**Marginal products:**
$$\overline{\text{mpk}} = \bar{A} \cdot (\overline{\text{ymarg}})^{1/\rho_c - 1} \cdot \alpha_K \cdot \bar{k}^{\rho_c - 1}$$

$$\overline{w} = \bar{A} \cdot (\overline{\text{ymarg}})^{1/\rho_c - 1} \cdot \alpha_L \cdot \bar{l}^{\rho_c - 1}$$

$$\overline{\text{mpoil}} = \bar{A} \cdot (\overline{\text{ymarg}})^{1/\rho_c - 1} \cdot \alpha_O \cdot \bar{o}^{\rho_c - 1}$$

**Labor supply calibration:**
$$\gamma = \frac{\overline{w}(1-\tau_w)\bar{\lambda}}{\bar{l}^{1/\phi_L}}$$

where $\bar{\lambda} = \frac{\bar{c}^{-\sigma_c}}{1+\tau_c}$

This ensures labor market equilibrium at the desired steady-state hours.

## Economic Interpretation

### Transmission Mechanisms

#### 1. **Positive TFP Shock** ($\varepsilon_t^a > 0$, $a_t$ increases)

The shock propagates through multiple channels:

**Direct production effect:**
- Output rises: $y_t = \bar{A} e^{a_t} \hat{y}_t \uparrow$
- All marginal products increase proportionally
- Factor income rises across the board

**Factor market responses:**
- Higher $\text{mpk}_t$ → investment becomes more profitable
- Higher $w_t$ → labor demand increases
- Higher $\text{mpoil}_t$ → oil demand increases

**Household behavior:**
- Income effect: Consumption rises ($c_t \uparrow$)
- Wealth effect: Labor supply may fall due to higher income
- Intertemporal substitution: Investment responds to higher $q_t$

**External sector:**
- Current account improves: $y_t \uparrow > (c_t + i_t + \text{oil imports})_t$
- Net foreign assets accumulate ($\text{nfa}_t \uparrow$)
- Foreign interest rate falls: $r_t = \bar{r} - \Phi_B \text{nfa}_{t-1} \downarrow$

**Nominal effects:**
- Marginal cost falls: $mc_t \downarrow$ (productivity gain)
- Inflation decreases via Phillips curve
- Central bank may lower interest rate in response

#### 2. **Oil Price Shock** ($\varepsilon_t^{oil} > 0$, $a_t^{oil}$ increases)

An adverse terms-of-trade shock with stagflationary effects:

**Cost effect:**
- Real cost of oil imports rises: $\frac{s_t}{p_t} \bar{P}^{oil} e^{a_t^{oil}} \uparrow$
- Oil demand adjusts: firms substitute toward capital and labor (limited by $\rho_c = -0.1$)

**Production impact:**
- With $\rho_c < 0$, oil is a **complement** to other inputs
- Effective output falls for given $k, l$ due to reduced oil usage
- Factor productivity declines

**Income and expenditure:**
- Real income falls (terms-of-trade loss)
- Consumption must decline: $c_t \downarrow$
- Investment may fall if profitability declines

**External balance:**
- Current account deteriorates: higher oil import bill
- Net foreign assets decline: $\text{nfa}_t \downarrow$
- Foreign interest rate rises: $r_t = \bar{r} - \Phi_B \text{nfa}_{t-1} \uparrow$

**Inflation dynamics:**
- Marginal cost rises: $mc_t \uparrow$ (cost-push)
- Inflation increases via Phillips curve: $\pi_t \uparrow$
- Central bank faces trade-off: stabilize inflation vs. output

**Monetary policy dilemma:**
- Raising $r_t^{dom}$ fights inflation but worsens output contraction
- Lowering $r_t^{dom}$ supports output but accommodates inflation
- Taylor rule parameters $(\phi_\pi, \phi_y)$ determine response

#### 3. **Interaction with Sticky Prices** ($\theta = 0.75$)

Nominal rigidities amplify real effects:

- Firms cannot fully adjust prices → quantities adjust instead
- Inflation responds gradually to cost changes
- Real interest rate movements affect intertemporal decisions
- Monetary policy becomes effective through real rate channel

**Phillips curve mechanism:**
$$\pi_t = \beta \pi_{t+1} + \underbrace{\frac{(1-\theta)(1-\beta\theta)}{\theta}}_{\text{slope}} \log\left(\frac{mc_{t-1}}{\overline{mc}}\right)$$

With $\theta = 0.75$, the slope is relatively flat → inflation is **persistent** but **sluggish**.

#### 4. **International Linkages**

**Net foreign asset dynamics:**
$$\text{nfa}_t = \underbrace{(1+r_t)\text{nfa}_{t-1}}_{\text{interest income}} + \underbrace{\text{CA}_t}_{\text{current account}}$$

where $\text{CA}_t = y_t - (1+\tau_c)c_t - i_t - \text{oil imports}_t - \text{tax revenues}_t$

**Debt-elastic interest rate:**
- Acts as a stabilizing force: prevents explosive NFA paths
- Higher debt → higher borrowing cost → incentive to save
- Parameter $\Phi_B = 0.001$ is small → gradual adjustment

**Uncovered interest parity:**
$$r_t^{dom} - r_t = \mathbb{E}_t[\Delta s_{t+1}]$$

- Links domestic and foreign interest rates
- Exchange rate adjusts to maintain equality
- Under perfect foresight: $\Delta s_{t+1}$ is known exactly

### Policy Implications

#### Monetary Policy Trade-offs

- **Inflation targeting** ($\phi_\pi = 1.5 > 1$): Satisfies Taylor principle, ensures determinacy
- **Output stabilization** ($\phi_y = 0.5$): Mitigates output fluctuations
- **Interest rate smoothing** ($\rho_r = 0.9$): Gradual adjustments prevent volatility

**Zero lower bound constraint:**
- Binds during severe deflationary episodes
- Limits monetary policy effectiveness
- May require unconventional policies (not in this model)

#### Fiscal Policy

Tax distortions affect steady state and dynamics:
- $\tau_c = 0.1$: Reduces consumption demand
- $\tau_w = 0.25$: Distorts labor supply decision
- $\tau_k = 0.15$: Reduces investment incentives

#### Structural Features

- **Capital adjustment costs** ($\Phi = 10$): Smooth investment responses, prevent excess volatility
- **CES with complements** ($\rho_c = -0.1$): Oil is essential → large responses to oil shocks
- **Debt-elastic premium** ($\Phi_B$): Prevents unlimited foreign borrowing

## Notes on Model Features

### CES Production with Input Complementarity

**Elasticity of substitution:**
$$\sigma_{CES} = \frac{1}{1-\rho_c} = \frac{1}{1-(-0.1)} = \frac{1}{1.1} \approx 0.91$$

Since $\sigma_{CES} < 1$, inputs are **complements** (less substitutable than in Cobb-Douglas):
- Oil is relatively **essential** to production
- Cannot easily substitute away from oil when prices rise
- Oil price shocks have **larger** real effects than in Cobb-Douglas case

**Limiting cases:**
- $\rho_c \to 0$: Cobb-Douglas ($\sigma_{CES} = 1$)
- $\rho_c \to -\infty$: Leontief/perfect complements ($\sigma_{CES} \to 0$)
- $\rho_c \to 1$: Linear/perfect substitutes ($\sigma_{CES} \to \infty$)

### Q-Theory of Investment

Investment follows Tobin's Q with quadratic adjustment costs:

$$C(k_{t-1}, i_t) = \frac{\Phi}{2}\left(\frac{i_t - \delta k_{t-1}}{k_{t-1}}\right)^2 k_{t-1}$$

**Implications:**
- Investment responds **gradually** to changes in profitability
- Smooths capital stock dynamics
- Prevents unrealistic investment volatility
- Parameter $\Phi = 10$ implies moderate adjustment costs

**First-order condition:**
$$1 = q_t - \Phi\left(\frac{i_t - \delta k_{t-1}}{k_{t-1}}\right)$$

When $q_t > 1$, firms invest above depreciation; when $q_t < 1$, they invest below.

### Debt-Elastic Interest Rate Premium

Prevents explosive debt dynamics:

$$r_t = \bar{r} - \Phi_B \cdot \text{nfa}_{t-1}$$

**Mechanism:**
- Net debtor ($\text{nfa} < 0$): Pays premium → $r_t > \bar{r}$
- Net creditor ($\text{nfa} > 0$): Receives lower rate → $r_t < \bar{r}$
- Ensures stationarity of NFA

**Parameter value:** $\Phi_B = 0.001$ is small → adjustment is **gradual**

### Calvo Price Stickiness

**Average price duration:**
$$\text{Duration} = \frac{1}{1-\theta} = \frac{1}{1-0.75} = 4 \text{ periods}$$

With $\theta = 0.75$:
- Each period, 75% of firms keep prices **unchanged**
- Each period, 25% of firms **re-optimize** prices
- Average firm adjusts price every 4 periods

**New Keynesian Phillips Curve slope:**
$$\kappa = \frac{(1-\theta)(1-\beta\theta)}{\theta} \approx \frac{(0.25)(0.25)}{0.75} \approx 0.083$$

The slope is **relatively flat** → inflation adjusts **slowly** to cost shocks.

### Forward-Looking Taylor Rule

$$r_t^{dom} = (1-\rho_r)\left[\bar{r} + \phi_\pi \mathbb{E}_t[\pi_{t+1}] + \phi_y \log\left(\frac{\mathbb{E}_t[y_{t+1}]}{\bar{y}}\right)\right] + \rho_r r_{t-1}^{dom}$$

**Key features:**
- Central bank responds to **expected** (not current) inflation and output
- Taylor principle satisfied: $\phi_\pi = 1.5 > 1$
- High smoothing: $\rho_r = 0.9$ → gradual rate changes
- Zero lower bound: $r_t^{dom} \geq 0$ imposed explicitly

**Rationale for forward-looking rule:**
- Monetary policy operates with lags
- Preemptive response to expected inflation
- More realistic for modern central banking

### Uncovered Interest Parity (UIP)

Under perfect foresight:
$$r_t^{dom} - r_t = s_{t+1} - s_t$$

**Exchange rate determination:**
- Higher domestic rate → currency **appreciates** today ($s_t$ falls)
- Expected depreciation offsets interest differential
- Links domestic monetary policy to exchange rate

**Note:** In stochastic simulations, this would require risk premium adjustments.

### Tax Distortions

The model includes three tax instruments:

$$\lambda_t = \frac{c_t^{-\sigma_c}}{1+\tau_c} \quad \text{(consumption tax wedge)}$$

$$l_t = \left(\frac{w_{t-1}(1-\tau_w)\lambda_t}{\gamma}\right)^{\phi_L} \quad \text{(labor tax wedge)}$$

$$\text{After-tax capital return} = (1-\tau_k) r_t^k \quad \text{(capital tax wedge)}$$

**Effects:**
- Reduce steady-state consumption, labor, and capital
- Distort intertemporal and intratemporal margins
- Generate tax revenue (subtracted from current account)

```julia

using Random
using CSV
using DataFrames

# Set parameters
T = 1000

# Set random seed for reproducibility
Random.seed!(1906)

# Create shock paths - standard normal (mean=0, std=1)
eps_a = [0 ;  randn(T)]
eps_oil = [0; randn(T)]

# Create DataFrame
shock_df = DataFrame(
   
    eps_a = eps_a,
    eps_oil = eps_oil
)

# Save to CSV
CSV.write("shocks.csv", shock_df, header = true)


println("\nFirst 10 periods:")
display(first(shock_df, 10))
```

```julia
irfs = context.results.model_results[1].irfs
keys(irfs)
```

```julia
y_eps_a   = irfs[:eps_a][:y]
nfa_eps_a = irfs[:eps_a][:nfa]
pi_eps_a  = irfs[:eps_a][:pi]
```

```julia
y_eps_oil   = irfs[:eps_oil][:y]
nfa_eps_oil = irfs[:eps_oil][:nfa]
pi_eps_oil  = irfs[:eps_oil][:pi]
```

```julia
T_irf = length(y_eps_a)
t = 0:T_irf-1
```

```julia
using Plots
default(linewidth=2)

p1 = plot(y_eps_a, title="GDP (y) – TFP shock", xlabel="Periods", legend=false)
p2 = plot(nfa_eps_a, title="NFA – TFP shock", xlabel="Periods", legend=false)
p3 = plot(pi_eps_a, title="Inflation (π) – TFP shock", xlabel="Periods", legend=false)

plot(p1, p2, p3, layout=(3,1), size=(700,600))
```

```julia
p1 = plot(y_eps_oil, title="GDP (y) – Oil shock", xlabel="Periods", legend=false)
p2 = plot(nfa_eps_oil, title="NFA – Oil shock", xlabel="Periods", legend=false)
p3 = plot(pi_eps_oil, title="Inflation (π) – Oil shock", xlabel="Periods", legend=false)

plot(p1, p2, p3, layout=(3,1), size=(700,600))
```

```julia
plot(y_eps_a, label="TFP shock", linewidth=2)
plot!(y_eps_oil, label="Oil shock", linewidth=2,
      title="GDP IRFs: TFP vs Oil", xlabel="Periods")
```
